```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import torch
from transformers import AutoTokenizer, T5ForConditionalGeneration
```

```python
model_path = '/kaggle/input/transformers/t5-large'
model      = T5ForConditionalGeneration.from_pretrained(model_path).cuda()
tokenizer  = AutoTokenizer.from_pretrained(model_path)
```

####  Bu kod, bir doğrulama veri setini kullanarak eğitilmiş bir modelin değerlendirilmesini gerçekleştirir. İşlem adımları şunlardır:

1. Veri seti okunur ve bir pandas DataFrame'e atılır.
2. Başlangıçta puan sıfırlanır.
3. Model değerlendirme moduna alınır.
4. Her bir veri örneği için aşağıdaki adımlar gerçekleştirilir:
	a. Veri örneğinin sütunları alınır.
	b. Kayıp puanları depolamak için bir liste oluşturulur.
	c. Giriş verisi tokenize edilir ve giriş ID'lerine dönüştürülür.
	d. Etiketler tokenize edilir ve gerekli işlemler uygulanır.
	e. Kayıp hesaplanır ve puan listesine eklenir.

5. Toplam puan, veri seti boyutuna bölünerek ortalama puan hesaplanır.

#### Bu şekilde, kod eğitilmiş bir modelin değerlendirilmesini gerçekleştirir ve ortalama puanı yazdırır.

```python
df_valid = pd.read_csv('/kaggle/input/kaggle-llm-science-exam/train.csv')
valid_score = 0
model.eval()
for index in tqdm(range(df_valid.shape[0])):
    columns = df_valid.iloc[index].values
    scores = []
    input_ids = tokenizer(columns[1]+" <extra_id_0>", return_tensors="pt").input_ids.cuda()
    labels = tokenizer(["<extra_id_0> "+columns[2+p] for p in range(5)], return_tensors="pt", padding=True).input_ids
    minlen = np.min([len(l) for l in labels])
    for p in range(5):
        with torch.no_grad():
            loss = model(input_ids=input_ids, labels=labels[p][:minlen].unsqueeze(0).cuda()).loss.detach().cpu().numpy()
        scores.append(float(loss))
    predict = np.array(list("ABCDE"))[np.argsort(scores)][:3].tolist()
    if columns[7] in predict:
        valid_score += [1,0.5,0.333333333333][predict.index(columns[7])]
valid_score /= df_valid.shape[0]
print(f'score = {valid_score}')
```

```python
from typing import Optional, Union
import pandas as pd
import numpy as np
import torch
from datasets import Dataset
from dataclasses import dataclass
from transformers import AutoTokenizer
from transformers.tokenization_utils_base import PreTrainedTokenizerBase, PaddingStrategy
from transformers import AutoModelForMultipleChoice, TrainingArguments, Trainer
```

#### This code allows a model trained for a multiple choice problem to generate predictions on the test dataset. The operation steps are as follows:

1. **options = 'ABCDE' and indices = list(range(5)):** Lists with options and indices are defined.

2. **option_to_index = {option: index for option, index in zip(options, indices)} and index_to_option = {index: option for option, index in zip(options, indices)}:** Dictionaries are created that associate options with indices.

3. Define the **preprocess function:** This function takes a sample (for example a line) and expects prompt as a dictionary with keys A, B, C, D, E and answer. The example copies the question 5 times and adds the answers before it is converted to BERT understandable token IDs. Finally, the label value is determined using the option_to_index dictionary.

4. Define the **DataCollatorForMultipleChoice class:** This class is used to manipulate dataset with multiple choices. In particular, the tokenizer, padding method, maximum length and other parameters are specified. When this class is called, the list of properties is retrieved and the options within each property are individually edited and combined. Then, using the tokenizer, the edited features are filled and tensors are created. Finally, a group is created and returned by combining the labels and tensors.

5. Define the **training_args variable:** This variable specifies the training arguments. Output index, evaluation strategy, save strategy, learning rate, total number of iterations, etc. contains values ​​such as

6. **Model and tokenizer instances are created:** The model is used to convert the tokenizer text into token IDs when loading a pre-trained model.

7. Create the **trainer object:** This object contains the model, training arguments, tokenizer, and data collector required to train the model.

8. The test dataset is read and dumped into a DataFrame.

9. The necessary preparatory steps are performed to generate the predictions on the test dataset: the answer column is set to 'A' and the data is preprocessed using the preprocess function.

10. Tokenized test dataset is created.

11. The model makes predictions on the test data set.

#### In this way, the code generates predictions of a trained model on the test dataset and keeps the prediction results in the variable test_predictions.

#### Bu kod, bir çoklu seçim problemi için eğitilmiş bir modelin test veri seti üzerinde tahminler oluşturmasını sağlar. İşlem adımları şu şekildedir:

1. **options = 'ABCDE' ve indices = list(range(5)):** Seçenekleri ve indisleri içeren listeler tanımlanır.

2. **option_to_index = {option: index for option, index in zip(options, indices)} ve index_to_option = {index: option for option, index in zip(options, indices)}:** Seçenekleri indislerle ilişkilendiren sözlükler oluşturulur.

3. **preprocess fonksiyonu tanımlanır:** Bu fonksiyon, bir örneği (örneğin bir satırı) alır ve prompt, A, B, C, D, E ve answer olmak üzere anahtarları olan bir sözlük olarak bekler. Örnek, BERT tarafından anlaşılabilir token ID'lerine dönüştürülmeden önce soruyu 5 kez kopyalar ve yanıtları ekler. Son olarak, etiket (label) değeri, option_to_index sözlüğü kullanılarak belirlenir.

4. **DataCollatorForMultipleChoice sınıfı tanımlanır:** Bu sınıf, birden çok seçenekli veri kümesini işlemek için kullanılır. Özellikle tokenizer, dolgu (padding) yöntemi, maksimum uzunluk ve diğer parametreler belirtilir. Bu sınıf çağrıldığında, özellikler listesi alınır ve her bir özelliğin içindeki seçenekler ayrı ayrı düzenlenerek birleştirilir. Ardından, tokenizer kullanılarak düzenlenmiş özelliklerin dolgu işlemi yapılır ve tensörler oluşturulur. Son olarak, etiketler ve tensörler birleştirilerek bir grup oluşturulur ve döndürülür.

5. **training_args değişkeni tanımlanır:** Bu değişken, eğitim argümanlarını belirtir. Çıktı dizini, değerlendirme stratejisi, kaydetme stratejisi, öğrenme oranı, toplam yineleme sayısı, vb. gibi değerler içerir.

6. **Model ve tokenizer örnekleri oluşturulur:** Model, önceden eğitilmiş bir modeli yüklerken, tokenizer metnin token ID'lerine dönüştürülmesi için kullanılır.

7. **trainer nesnesi oluşturulur:** Bu nesne, modeli eğitmek için gerekli olan model, eğitim argümanları, tokenizer ve veri toplayıcısını içerir.

8. Test veri seti okunur ve bir DataFrame'e atılır.

9. Test veri seti üzerinde tahminlerin oluşturulması için gerekli hazırlık adımları gerçekleştirilir: answer sütunu 'A' olarak ayarlanır ve preprocess fonksiyonu kullanılarak veri önişleme yapılır.

10. Tokenize edilmiş test veri kümesi oluşturulur.

11. Model tarafından test veri seti üzerinde tahminler yapılır.

#### Bu şekilde, kod eğitilmiş bir modelin test veri seti üzerinde tahminlerini oluşturur ve tahmin sonuçlarını test_predictions değişkeninde tutar.



```python
options = 'ABCDE'
indices = list(range(5))

option_to_index = {option: index for option, index in zip(options, indices)}
index_to_option = {index: option for option, index in zip(options, indices)}

def preprocess(example):
    """The example is expected to be a dictionary with keys 'prompt', 'A', 'B', 'C', 'D', 'E', and 'answer'."""
    # The AutoModelForMultipleChoice class expects a set of question/answer pairs
    # so we'll copy our question 5 times before tokenizing
    first_sentence = [example['prompt']] * 5
    second_sentence = [example[option] for option in options]
    # Our tokenizer will turn our text into token IDs BERT can understand
    tokenized_example = tokenizer(first_sentence, second_sentence, truncation=True)
    tokenized_example['label'] = option_to_index[example['answer']]
    
    return tokenized_example

@dataclass
class DataCollatorForMultipleChoice:
    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None
    
    def __call__(self, features):
        label_name = 'label' if 'label' in features[0].keys() else 'labels'
        labels = [feature.pop(label_name) for feature in features]
        batch_size = len(features)
        num_choices = len(features[0]['input_ids'])
        flattened_features = [
            [{k: v[i] for k, v in feature.items()} for i in range(num_choices)] for feature in features
        ]
        flattened_features = sum(flattened_features, [])
        
        batch = self.tokenizer.pad(
            flattened_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors='pt',
        )
        batch = {k: v.view(batch_size, num_choices, -1) for k, v in batch.items()}
        batch['labels'] = torch.tensor(labels, dtype=torch.int64)
        return batch   
        

training_args = TrainingArguments(
    output_dir='./',
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    learning_rate=3e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    num_train_epochs=10,
    weight_decay=0.01,
    report_to='none'
)

# Generally it's a bad idea to validate on your training set, but because our training set
# for this problem is so small we're going to train on all our data.

model_dir = '/kaggle/input/llm-sci-exam-deberta-large-run01'
tokenizer = AutoTokenizer.from_pretrained(model_dir)
model = AutoModelForMultipleChoice.from_pretrained(model_dir)

trainer = Trainer(
    model=model,
    args=training_args,
    tokenizer=tokenizer,
    data_collator=DataCollatorForMultipleChoice(tokenizer=tokenizer),
)

test_df = pd.read_csv('/kaggle/input/kaggle-llm-science-exam/test.csv')
test_df.head()

# There are more verbose/elegant ways of doing this, but if we give our test set a random `answer` column
# we can make predictions directly with our trainer.
test_df['answer'] = 'A'

# Other than that we'll preprocess it in the same way we preprocessed test.csv
test_ds = Dataset.from_pandas(test_df)
tokenized_test_ds = test_ds.map(preprocess, batched=False, remove_columns=['prompt', 'A', 'B', 'C', 'D', 'E', 'answer'])

# Here we'll generate our "real" predictions on the test set
test_predictions = trainer.predict(tokenized_test_ds)
```

#### This code combines the forecast results to create the final forecasts and saves them to a CSV file. The operation steps are as follows:
