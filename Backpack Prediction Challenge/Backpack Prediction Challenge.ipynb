In the last days of this competition we will try to "divide attention.." The idea is based on and consists of the following:

The intersection of two beautiful notebooks - of the American grandmaster [Chris Deotte](https://www.kaggle.com/cdeotte/code) and the Japanese expert [Masaya Kawamata](https://www.kaggle.com/masayakawamata/code) - each of which respectively have ratings in our leaderboard as: lb.**1** = [**38.84716**](https://www.kaggle.com/code/cdeotte/feature-engineering-with-rapids-lb-38-847) and lb.**2** = [**38.84771**](https://www.kaggle.com/code/masayakawamata/s5e2-sae-te-xgb). Together 50**×**50 they showed lb.**1**+**2** = [**38.84293**](https://www.kaggle.com/code/vyacheslavbolotin/ps-s5e2-ensemble-of-solutions?scriptVersionId=223931636). Also on the histograms, those at the end of the [ensemble notebook](https://www.kaggle.com/code/vyacheslavbolotin/ps-s5e2-ensemble-of-solutions), version [83](https://www.kaggle.com/code/vyacheslavbolotin/ps-s5e2-ensemble-of-solutions?scriptVersionId=223931636), and correspond to their "friendly" intersection, it is clear that Chris Deotte's solution "looks" a little wider. And if they complemented each other, and clarified each other, we can assume that Masaya Kawamata's solution should be at some of the depths a little better. We decided to conduct a simple experiment - to divide their attention, that is, somewhere they will also work 50**×**50. And somewhere, well, let's say 65**×**35. We mean percent. How to find this "somewhere". That's the question **??**.
Therefore, we will do the following: Everything that concerns the middle - this is, for example, 79-81, there they will work 50**×**50. And where it is higher, 65**×**35. Everything is clear with the middle. It is not clear with the other two parts - the top and the bottom. Therefore, we will do two experiments, alternately giving each of the two parts according to each of the solutions of this tandem. Perhaps the boundaries should be expanded or narrowed, and the parts related to the top and bottom should be left for consideration only by one of the solutions. If this idea and this approach "enlightens" something, then a powerful mechanism based on ML can be turned on. Also, not the solutions themselves will be used for this, but their experienced and already evaluated results. Which can be easily restored by launching these [excellent solutions](https://www.kaggle.com/competitions/playground-series-s5e2/code?competitionId=90274&sortBy=scoreAscending&excludeNonAccessedDatasources=true).
```python
def visual_distribution_target(target, df, color="#777abc", title='target'):
    import matplotlib.pyplot as plt
    _title = f"{target} Distribution" if title == 'target' else title
    plt.figure(figsize=(15, 4))
    df[target].plot(color=color, linewidth=0.1, title=_title)
    plt.axhline(df[target].mean(), c="w", ls="-", lw=2.5)
    plt.show()


def preparing_submission(v, df, orient='mean',display_df=False):
    V = v
    df[" -- 50 x 50 -- "]  = df.apply(lambda x: (x["subm_CD"] + x["subm_MK"])/2, axis=1)
    df["divide_attention"] = df.apply(lambda x: divide_attention(x,V,orient), axis=1)
    df["50x50 - div_att"]  = df.apply(lambda x: x[" -- 50 x 50 -- "] - x["divide_attention"], axis=1)
    if display_df: display(df.head(7))
```

```python

def divide_attention(x,V,orient):

    if orient == 'mean': orient = (x["subm_CD"] + x["subm_MK"]) / 2
    if orient == 'sub1': orient =  x["subm_CD"]
    if orient == 'sub2': orient =                 x["subm_MK"]

    
    if V==1 or V==2 or V==3:
        if orient >= 79 and orient <= 81: return orient
        else:
            if V==1:
                if orient < 79: return x["subm_CD"] *0.35 + 0.65* x["subm_MK"]
                if orient > 81: return x["subm_MK"] *0.35 + 0.65* x["subm_CD"]
            elif V==2:
                if orient < 79: return x["subm_MK"] *0.35 + 0.65* x["subm_CD"]
                if orient > 81: return x["subm_CD"] *0.35 + 0.65* x["subm_MK"]
            elif V==3:
                if orient < 79: return x["subm_CD"] *0.65 + 0.35* x["subm_MK"]
                if orient > 81: return x["subm_CD"] *0.65 + 0.35* x["subm_MK"]
                
    if V==4:
        if orient >= 78.5 and orient <= 81.5: return orient
        else:
            if orient < 78.5: return x["subm_MK"] *0.40 + 0.60* x["subm_CD"]
            if orient > 81.5: return x["subm_CD"] *0.40 + 0.60* x["subm_MK"]
                
    if V==5 or V==6 or V==7:
        if orient >= 80 and orient <= 83: return orient
        else:   
            if orient < 80: return x["subm_MK"] *0.45 + 0.55* x["subm_CD"]
            if orient > 83: return x["subm_MK"] *0.45 + 0.55* x["subm_CD"]

    if V==8:
        if orient >= 80 and orient <= 83: return x["subm_CD"]*0.30 + 0.70*x["subm_MK"]
        else:   
            if orient < 80: return x["subm_MK"] *0.30 + 0.70* x["subm_CD"]
            if orient > 83: return x["subm_MK"] *0.30 + 0.70* x["subm_CD"]
                
    if V==9:
        if orient >= 80 and orient <= 83: return x["subm_CD"]*0.15 + 0.85*x["subm_MK"]
        else:   
            if orient < 80: return x["subm_MK"] *0.44 + 0.56* x["subm_CD"]
            if orient > 83: return x["subm_MK"] *0.44 + 0.56* x["subm_CD"]
                
# orient - scipy.stats.rankdata
```
